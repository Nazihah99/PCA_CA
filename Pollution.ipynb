```python
# Select the relevant attributes for clustering
attributes = ['PRECReal', 'JANTReal', 'JULTReal', 'OVR65Real', 'POPNReal', 'EDUCReal', 'HOUSReal', 'DENSReal', 'NONWReal',
              'WWDRKReal', 'POORReal', 'HCReal', 'NOXReal', 'SO@Real', 'HUMIDReal', 'MORTReal']
X = data[attributes]
```

```python
# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Calculate PCA with different numbers of components
n_components = np.arange(1, len(X_scaled[0])+1)
explained_variance = []
cumulative_variance = []
silhouette_scores = []
wcss = []

for n in n_components:
    pca = PCA(n_components=n)
    pca.fit(X_scaled)
    transformed_data = pca.transform(X_scaled)

    # Perform clustering with different numbers of clusters
    cluster_range = range(2, 10)  # Adjust the range as needed
    silhouette_scores_per_cluster = []
    wcss_per_cluster = []

    for k in cluster_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(transformed_data)
        labels = kmeans.labels_
        silhouette_scores_per_cluster.append(silhouette_score(transformed_data, labels))
        wcss_per_cluster.append(kmeans.inertia_)

    # Store the results for each number of components
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
    cumulative_variance.append(np.sum(pca.explained_variance_ratio_))
    silhouette_scores.append(silhouette_scores_per_cluster)
    wcss.append(wcss_per_cluster)

# Plot explained variance ratio
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(n_components, explained_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.savefig('scree_plot.png')

# Plot cumulative explained variance ratio
plt.subplot(1, 2, 2)
plt.plot(n_components, cumulative_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Cumulative Explained Variance Plot')
plt.savefig('cumulative_variance_plot.png')

# Plot silhouette scores and WCSS for each number of components
for i, n in enumerate(n_components):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(cluster_range, silhouette_scores[i], marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Score')
    plt.title(f'Silhouette Scores for PCA with {n} Components')
    plt.savefig(f'silhouette_scores_pca_{n}_components.png')

    plt.subplot(1, 2, 2)
    plt.plot(cluster_range, wcss[i], marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('WCSS')
    plt.title(f'WCSS for PCA with {n} Components')
    plt.savefig(f'wcss_pca_{n}_components.png')

plt.show()
```

```python
# Apply PCA
pca = PCA(n_components=4)
principal_components = pca.fit_transform(X_scaled)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

wcss = []
wcsspca = []
silhouette_scores = []
silhouette_scorespca = []

# Elbow Method
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

for m in range(1, 10):
    kmeanspca = KMeans(n_clusters=m, random_state=42)
    kmeanspca.fit(principal_components)
    wcsspca.append(kmeanspca.inertia_)

# Silhouette Score
for j in range(2, 10):
    kmeans2 = KMeans(n_clusters=j, random_state=42)
    kmeans2.fit(X_scaled)
    silhouette_scores.append(silhouette_score(principal_components, kmeans2.labels_))
    
for p in range(2, 10):
    kmeanspca2 = KMeans(n_clusters=p, random_state=42)
    kmeanspca2.fit(principal_components)
    silhouette_scorespca.append(silhouette_score(principal_components, kmeanspca2.labels_))

# Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Elbow Method
axs[0, 0].plot(range(1, 10), wcss)
axs[0, 0].set_xlabel('Number of clusters (k)')
axs[0, 0].set_ylabel('WCSS')
axs[0, 0].set_title('Elbow Method without PCA')

# Plot 2: Silhouette Score
axs[0, 1].plot(range(2, 10), silhouette_scores)
axs[0, 1].set_xlabel('Number of clusters (k)')
axs[0, 1].set_ylabel('Silhouette Score')
axs[0, 1].set_title('Silhouette Score without PCA')

# Plot 3: (Empty plot)
axs[1, 0].plot(range(1, 10), wcsspca)
axs[1, 0].set_xlabel('Number of clusters (k)')
axs[1, 0].set_ylabel('WCSS')
axs[1, 0].set_title('Elbow Method with PCA')

# Plot 4: (Empty plot)
axs[1, 1].plot(range(2, 10), silhouette_scorespca)
axs[1, 1].set_xlabel('Number of clusters (k)')
axs[1, 1].set_ylabel('Silhouette Score')
axs[1, 1].set_title('Silhouette Score with PCA')

# Adjust spacing between subplots
plt.tight_layout()

# Save the figure
plt.savefig('my_figure.png')

# Display the figure
plt.show()

```
